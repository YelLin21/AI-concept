```python
import torch

# Create a 2D random tensor
x1 = torch.rand((5, 6))

#the shape, data type, and mean of all elements
x1_shape = x1.shape
x1_dtype = x1.dtype
x1_mean = x1.mean().item()

print(f"Shape of x1: {x1_shape}")
print(f"Data type of x1: {x1_dtype}")
print(f"Mean of all elements in x1: {x1_mean}")
```

2.2 Assignment 2

```python
import torch
x2 = torch.randint(0, 10, (4, 4))

x2_float = x2.float()
difference = x2_float - x2
print("x2 (Integer Tensor):")
print(x2)
print("\nx2_float (Converted to Float):")
print(x2_float)
print("\nDifference (x2_float - x2):")
print(difference)
```

The difference will be zero for all elements because the conversion of x2 from integer to float does not alter the underlying values, only their data type. The original values remain unchanged, and the subtraction results in zero.

```python
import torch

# Create two tensors x4_A and x4_B of shape (2, 3) with random integers in range [1, 5]
x4_A = torch.randint(1, 6, (2, 3))
x4_B = torch.randint(1, 6, (2, 3))

# Element-wise addition
addition_result = x4_A + x4_B

# Element-wise subtraction
subtraction_result = x4_A - x4_B

# Element-wise multiplication
multiplication_result = x4_A * x4_B

# Print each result and their shapes
print("x4_A:")
print(x4_A)
print("\nx4_B:")
print(x4_B)
print("\nAddition Result:")
print(addition_result)
print("Shape:", addition_result.shape)
print("\nSubtraction Result:")
print(subtraction_result)
print("Shape:", subtraction_result.shape)
print("\nMultiplication Result:")
print(multiplication_result)
print("Shape:", multiplication_result.shape)
```

```python
import torch

x6 = torch.rand((4, 4))
column_wise_sum = x6.sum(dim=0)
row_wise_sum = x6.sum(dim=1)

print("Tensor x6:")
print(x6)
print("\nColumn-wise sum:")
print(column_wise_sum)
print("Shape:", column_wise_sum.shape)
print("\nRow-wise sum:")
print(row_wise_sum)
print("Shape:", row_wise_sum.shape)
```

```python
import torch
x7_A = torch.rand((2, 2))
x7_B = torch.rand((2, 2))

concat_dim0 = torch.cat((x7_A, x7_B), dim=0)
concat_dim1 = torch.cat((x7_A, x7_B), dim=1)

print("Tensor x7_A:")
print(x7_A)
print("\nTensor x7_B:")
print(x7_B)
print("\nConcatenated along dimension 0 (Shape (4, 2)):")
print(concat_dim0)
print("\nConcatenated along dimension 1 (Shape (2, 4)):")
print(concat_dim1)
```

```python
import torch
x8 = torch.randint(0, 10, (2, 3, 4))
x8 = x8.float()
max_value = x8.max()
mean_dim2 = x8.mean(dim=2)

print("Tensor x8:")
print(x8)
print("\nMaximum value in x8:", max_value)
print("\nMean along dimension 2:")
print(mean_dim2)
print("Shape of mean along dimension 2:", mean_dim2.shape)
```
